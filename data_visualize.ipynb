Add library

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")

print("Tải thư viện thành công!")
```

Loading Data

```python
df = pd.read_csv('dataset/songs_normalize.csv')

print("Tải dữ liệu thành công. Dữ liệu có", df.shape[0], "hàng và", df.shape[1], "cột.")
df.head()
```

Define "Hit"

```python
print(df['popularity'].describe())

plt.figure(figsize=(12, 6))
sns.histplot(df['popularity'], bins=50, kde=True)
plt.title('Phân phối của Popularity (Độ nổi tiếng)')
plt.xlabel('Popularity (0-100)')
plt.ylabel('Số lượng bài hát')
plt.axvline(x=80, color='red', linestyle='--', label='Ngưỡng "Hit" đề xuất (80)')
plt.legend()
plt.show()
```

```python
NGUONG_HIT = 80

df['is_hit'] = (df['popularity'] > NGUONG_HIT).astype(int)

# Kiểm tra xem dữ liệu có bị mất cân bằng không
print("Phân bố các lớp 'Hit' (1) và 'Not Hit' (0):")
print(df['is_hit'].value_counts(normalize=True))
```
